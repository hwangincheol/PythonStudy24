함수 (function)

함수는 단일 동작을 수행하는 블럭을 말한다.
자바에서는 {} 블럭 단위로 수행하는데 파이썬은 인터프리터로 줄단위 수행

프로그램의 메모리 영역은
크게 3가지로 나뉘어짐

스택영역 : 변수에 기본타입이나 참조타입을 연결
힙영역 : 객체를 생성해서 주소를 가지고 있음
메서드영역 : 동작들을 보관하고 있다가 호출되면 매개값(파라미터)을 처리 후 리턴을 제공 (함수나, 메서드를 보관)

함수를 사용하는 이유 : 반복적인 코드 발생을 최소화 화기 위해서 (CRUD)


```python
# 파이썬 함수의 구조
# def (define : 정의) 함수 생성
# def 함수이름(매개변수) :
#     실행문1 (4칸띄우기)
#     실행문2
#     ...

def add(a, b):      # add라는 함수를 정의(생성), 매개값은 2개를 받음
    print("add() 메서드 종료")  # return이 없을때는 프린트로 종료 찍어주는게 좋음
    return a + b    # 매개값으로 받은 2개의 변수를 더해 리턴
    # add() 매서드 종료

print(add(4, 5))
print(add("황", "인철"))
```

```python
# *args : 매개값이 몇개인지 모를때 (1개 이상~)

def add_many(*args): # *args 매개값이 1개 이상!
    result = 0  # 결과를 저장하는 변수
    for i in args:  # *args는 리스트로 저장이 됨
        result += i # result = result + i
    return result   # 결과 리턴

print(add_many(1, 2, 3, 4, 5, 6, 7, 8, 9))
# args는 관례임(프로그래머들끼리의 약속)
```

```python
test = add_many(9, 8, 7, 6)
```

```python
def add_mul(choice, *args):
    if choice == "add":
        result = 0
        for i in args:
            result += i
    elif choice == "mul":
        result = 1
        for i in args:
            result *= i
    return result
```

```python
result = add_mul("add", 1,2,3,4,5)
print(result)
```

```python
result = add_mul("mul", 3, 4, 5)
print(result)
```

```python
result = add_mul("add", "황", "인", "철")   # result가 문자로 들어가서 오류
print(result)
```

```python
# 람다식 : 인터페이스 형태의 실행문이 1개인 코드
# 자바에서 람다식은 -> 사용하고 자바스크립트 =>
# 파이썬의 람다식은
# 변수 = lambda 매개변수 : 실행문
add = lambda a, b : a + b
result = add(3, 4)
print(result)
```

```python
# 미션
# 자판기 코드를 함수로 리빌드 진행
# 커피 종류가 5개 (커피명, 수량, 단가)
# 사용자 커피를 반복 구매
# 관리자가 판매종료 후 통계

## 기초 데이터 (전역 변수)
drink = [[" ", 0, 0] for _ in range(5)] # 음료 배열 5개 생성 (이름, 수량, 가격)
totalMoney = 0  # 넣은 금액
sumPrice = 0    # 판매 총액

## 함수
# 자판기 처음 음료 5개 채워 넣기 함수
def create():
    print(">> 총 5개 음료의 이름, 수량, 가격을 입력해주세요 <<")
    global drink
    for a in range(0,5):
        drink[a][0] = input(f"{a + 1}번째 음료 이름을 입력해주세요 : ")
        while True:
            drink[a][1] = input(f"{a + 1}번째 음료 수량을 입력해주세요 : ")
            if drink[a][1].isdigit():
                drink[a][1] = int(drink[a][1])
                break
            else:
                print("수량은 숫자만 입력 가능합니다. 다시 입력해주세요.")
        while True:
            drink[a][2] = input(f"{a + 1}번째 음료 가격을 입력해주세요 : ")
            if drink[a][2].isdigit():
                drink[a][2] = int(drink[a][2])
                break
            else:
                print("가격은 숫자만 입력 가능합니다. 다시 입력해주세요.")
        print(f" >> {a + 1}. {drink[a][0]}, {drink[a][1]}개, {drink[a][2]}원\n")
    print("\n\n>> 입력이 완료 되었습니다. 영업을 시작합니다. <<")
# create() end

# 자판기 음료 수정
def update(x):
    global drink
    print("\n\n\n============================================================")
    print(f"              {drink[x][0]} 를 수정합니다.")
    print("============================================================")

    drink[x][0] = input("\n 새로운 음료명을 입력해주세요 : ")
    while True:
        drink[x][1] = input("\n 새로운 수량을 입력해주세요 : ")
        if drink[x][1].isdigit():
            drink[x][1] = int(drink[x][1])
            break
        else:
            print("수량은 숫자만 입력 가능합니다. 다시 입력해주세요.")
    while True:
        drink[x][2] = input("\n 새로운 가격을 입력해주세요 : ")
        if drink[x][2].isdigit():
            drink[x][2] = int(drink[x][2])
            break
        else:
            print("가격은 숫자만 입력 가능합니다. 다시 입력해주세요.")
    print(f" >> {x + 1}. {drink[x][0]}, {drink[x][1]}개, {drink[x][2]}원\n")
    print(" >> 수정이 완료 되었습니다.")
#update() end

# 구매
def buy(i):
    global drink, totalMoney, sumPrice
    if totalMoney >= drink[i][2]:
        if drink[i][1] > 0:
            drink[i][1] -= 1
            totalMoney = totalMoney - drink[i][2]
            sumPrice = sumPrice + drink[i][2]
            print(f">>> {drink[i][0]}을 구매합니다.")
        else: print(f">>> {drink[i][0]}은 현재 품절입니다.")
    else: print(">>> 금액이 부족합니다.")
# buy() end


## 프로그램 시작
print("== 자판기 프로그램을 시작합니다 ==")
create()    # 음료 생성

while True: # 자판기 가동
    prompt1 = f"""
    - 금액 투입 단계-
    =======================================
    | 1. {drink[0][0]} {drink[0][2]}원   | 2. {drink[1][0]} {drink[1][2]}원   |
    =======================================
    | 3. {drink[2][0]} {drink[2][2]}원   | 4. {drink[3][0]} {drink[3][2]}원   |
    =======================================
    | 5. {drink[4][0]} {drink[4][2]}원   |                  |
    =======================================

    구매하기전 원하는 만큼 금액을 투입하고
            '구매'를 입력해주세요.
    돈을 회수 하려면 '반환'을 입력해주세요.
    =======================================
    """

    prompt2 = f"""
    - 음료 구매 단계 -
    =======================================
    | 1. {drink[0][0]} {drink[0][2]}원   | 2. {drink[1][0]} {drink[1][2]}원   |
    =======================================
    | 3. {drink[2][0]} {drink[2][2]}원   | 4. {drink[3][0]} {drink[3][2]}원   |
    =======================================
    | 5. {drink[4][0]} {drink[4][2]}원   |                  |
    =======================================

        구매가 끝나셨으면 거스름 돈
            '반환'을 입력해주세요.
    =======================================
    """

    print("\n\n")
    print(prompt1)
    print(" >>>    현재 잔액 : %d원" % totalMoney)
    money = input("\n 얼마를 넣을까요? : ")
    if money == "종료":
        print("\n\n\n 프로그램을 종료합니다.")
        break
    elif money == "관리자":
        while True:
            print("\n\n\n===============================================================")
            print("                  관리자 모드를 실행합니다 ")
            print("                >>> 판매 총액 : %d원<<<" % sumPrice)
            print("===============================================================")
            print(f" 1.{drink[0][0]}({drink[0][1]}개), 2.{drink[1][0]}({drink[1][1]}개), 3.{drink[2][0]}({drink[2][1]}개), 4.{drink[3][0]}({drink[3][1]}개), 5.{drink[4][0]}({drink[4][1]}개)")
            print("===============================================================")
            print("| 1 ~ 5. 수정하고싶은 음료의 번호를 눌러주세요  | 0. 돌아가기 |")
            print("===============================================================")
            input2 = input("\n원하시는 메뉴를 선택해주세요 : ")
            if input2 == '0': break
            elif input2 == '1' or input2 == '2' or input2 == '3' or input2 == '4' or input2 == '5':
                x = int(input2) - 1
                update(x)   # 수정
            else: print("\n\n\n!!!!!! 잘못 입력하셨습니다. !!!!!!")
    elif money == "반환":
                print("\n\n\n 거스름돈 %d원을 반환합니다." % totalMoney)
                print(" 이용 해주셔서 감사합니다 ^^")
                totalMoney = 0
    elif money == "구매":
        while True:
            if totalMoney == 0:
                print("\n >>> 구매를 원하시면 잔액을 넣어주세요.")
                break
            print("\n\n")
            print(prompt2)
            print(" >>>    현재 잔액 : %d원" % totalMoney)
            input1 = input("\n 구매하고 싶은 음료의 번호를 입력해주세요 : ")
            print("\n\n\n >>> ")
            if input1 == "반환":
                print(" 거스름돈 %d원을 반환합니다." % totalMoney)
                print(" 이용 해주셔서 감사합니다 ^^")
                totalMoney = 0
                break
            elif input1 == '1' or input1 == '2' or input1 == '3' or input1 == '4' or input1 == '5':
                i = int(input1) - 1
                buy(i)  # 구매
            else: print("\n\n\n!!!!!! 잘못 입력하셨습니다. !!!!!!")
    elif not money.isdigit(): print("\n\n\n!!!!!! 잘못 입력하셨습니다. !!!!!!")
    else : totalMoney = totalMoney + int(money)
```
